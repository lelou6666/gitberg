```python
import csv
import sys
sys.path.append("..")
from gitenberg.book import Book

pg_ids={}
dups=set()
with open('../gitenberg/data/GITenberg_repo_list.tsv','r') as f:
    for vals in csv.reader(f,delimiter='\t', quotechar='"'):
        if int(vals[0]) in pg_ids:
            prev=pg_ids[int(vals[0])]
            pg_ids[int(vals[0])] = [prev,vals[1]] if isinstance(prev,str) else prev + [vals[1]]
            dups.add(int(vals[0]))
        else:
            pg_ids[int(vals[0])] = vals[1]


    for key in dups:
        newest = None
        for reponame in pg_ids[key]:
            book= Book(None,reponame)
            repo = book.github_repo.github.repository('GITenberg', reponame)
            if newest is not None and repo is not None:
                if newest.pushed_at > repo.pushed_at:
                    deletedrepo = repo
                else:
                    deletedrepo = newest
                    newest = repo
                print 'deleting {} pushed at {}'.format(deletedrepo.name,deletedrepo.pushed_at)
                deletedrepo.delete()
            elif repo is not None:
                newest = repo
        pg_ids[key] = newest.name
        
        
with open('../assets/GITenberg_repo_list_out.tsv','r+w') as out:
    csv.writer(out,delimiter='\t', quotechar='"').writerows([(key,pg_ids[key]) for key in pg_ids.keys()])
```
