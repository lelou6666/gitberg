```python
import csv
import sys
sys.path.append("..")
from gitenberg.push import GithubRepo

removed=[]
gh_handler=GithubRepo(None)
with open('../assets/to_remove_list.tsv','r') as f:
    for (pg_id, reponame, pg_type) in csv.reader(f,delimiter='\t', quotechar='"'):
        try:
            repo = gh_handler.github.repository('GITenberg', reponame)
            print 'deleting {}'.format(repo.name)
            repo.delete()
            removed.append([pg_id,reponame,pg_type])
        except:
            print 'error removing {}'.format(reponame)
            break
print '{} repos have been removed'.format(len(removed))
with open('../assets/removed.tsv','wb') as f:
    writer=csv.writer(f,delimiter='\t', quotechar='"')
    writer.writerows(removed)
```

```python
import csv

removed_ids=set()
with open('../assets/removed.tsv','r') as f:
    for (pg_id, reponame, pg_type) in csv.reader(f,delimiter='\t', quotechar='"'):
        removed_ids.add(pg_id)
with open('../assets/GITenberg_repo_list_out.tsv','wb') as out:
    writer=csv.writer(out,delimiter='\t', quotechar='"')
    with open('../gitenberg/data/GITenberg_repo_list.tsv','r') as f:
        for (pg_id, reponame) in csv.reader(f,delimiter='\t', quotechar='"'):
            if pg_id not in removed_ids:
                writer.writerow([pg_id, reponame])
```
